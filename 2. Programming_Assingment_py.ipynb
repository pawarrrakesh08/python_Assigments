### 1. Write a Python program to convert kilometers to miles?

```python
# 1 KM  == 0.621371 miles

#User input for the km

km = float(input("Enter the km value: "))

miles_val = 0.621371

km_to_miles = km * miles_val
print(km, "kilometers is equal to miles",km_to_miles)
```

### 4. Write a Python program to solve quadratic equation?


```python
# import complex math module
# j represent the complex number in python. it has value root-1
import cmath

print("General Form:- ax**2+bx+x = 0")

a = int(input("Enter a (a!=0): "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

#Calulation of Discriminant

d = (b**2)-(4*a*c)

sol1 = (-b-cmath.sqrt(d))/(2*a)
sol2 = (-b+cmath.sqrt(d))/(2*a)

print("\n")
print(f"Results for equation, {a}x**2+{b}x+{c} = 0, are:- \n")

if d > 0:
    print("Type of Roots : Two Distinct Real roots")
elif d == 0:
    print("Type of Roots : Two Real Equal roots")
elif d < 0:
    print("Type of Roots : Two Complex roots")

print(f"The Solutions are {sol1} and {sol2}")
```
